# Correlation Map

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create data
data = {
    'Cities': ['Houston', 'Dallas', 'Fort Worth', 'Austin', 'San Antonio'],
    '% Virally Suppressed': [57.6, 56.8, 63.4, 66.7, 59.8],
    '% Frequent Physical Distress': [11, 11, 11, 10, 14],
    '% Frequent Mental Distress': [11, 11, 11, 10, 11],
    'Food Insecurity': [17, 18, 17, 16, 11],
    'Limited Access to Healthy Food': [6, 7, 8, 7, 11],
    'Household Income': [56415, 54429, 61553, 70068, 53170],
    'Firearm Fatalities': [1033, 1426, 1001, 452, 2779],
    'Homicide Rate': [13, 11, 10, 8, 11]
}

df = pd.DataFrame(data)

# Correlation matrix
corr_matrix = df.drop(columns='Cities').corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 cmap='Blues',
                 annot_kws={"weight": "bold", "fontsize": 12})

# Customize axis labels
ax.set_xticklabels(ax.get_xticklabels(), fontweight='bold', fontsize=12, rotation=30, ha='right')
ax.set_yticklabels(ax.get_yticklabels(), fontweight='bold', fontsize=12, rotation=0)

# Add title
plt.title('Correlation Matrix - VS among BWLH and Selected CHR-R SDoH',
          fontweight='bold', fontsize=14)

plt.tight_layout()
plt.show()
```

# Bar and Line Chart

```python
# Create new data frame 
plot_df = data[['County', 'BWLH', '% Frequent Physical Distress', '% Frequent Mental Distress']].copy()
plot_df = plot_df.sort_values(by='BWLH', ascending=False)

# Extract sorted columns
counties = plot_df['County']
total_bwlh = plot_df['BWLH']
frequent_physical_distress = plot_df['% Frequent Physical Distress']
frequent_mental_distress = plot_df['% Frequent Mental Distress']

fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for Total BWLH 
bars = ax1.bar(counties, total_bwlh, color='red', label='BWLH')
ax1.set_xlabel('County', fontsize=12, fontweight='bold')
ax1.set_ylabel('BWLH', fontsize=12, fontweight='bold')
ax1.tick_params(axis='y', colors='black', labelsize=10)
ax1.set_xticklabels(counties, rotation=45, fontsize=10, fontweight='bold')

# Annotate bars with total BWLH values
for bar, total in zip(bars, total_bwlh):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, height, f'{int(total)}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Twin axis for line plot
ax2 = ax1.twinx()
ax2.plot(counties, frequent_physical_distress, color='blue', marker='o', linestyle='-', label='% Frequent Physical Distress')
ax2.plot(counties, frequent_mental_distress, color='green', marker='o', linestyle='-', label='% Frequent Mental Distress')
ax2.set_ylabel('% Distress', fontsize=12, fontweight='bold')
ax2.tick_params(axis='y', colors='black', labelsize=10)

# Combine legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(0.5, 0.85))

plt.title('Total Black Women Living with HIV (BWLH)and Mental and Physical Distress Levels by County', fontsize=14, fontweight='bold')
plt.tight_layout() 
plt.show()
```
